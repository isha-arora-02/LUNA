```python
pheno_sub = (
    pheno[pheno["disease_status"].isin(caps)]
      .groupby("disease_status", group_keys=False)
      .apply(lambda g: g.sample(n=min(len(g), caps[g.name]), random_state=42))
).reset_index(drop = True)
pheno_sub.to_csv("/labs/khatrilab/provido/luna_final/LUNA-main/gene_expr_data/final_pheno.csv")
pheno_sub
```

```python
order = pheno_sub["geo_accession"].tolist()

expr_sub = (
    expr.set_index("Unnamed: 0")   # GSM id column in expr
        .loc[order]                # reorder + subset to pheno_sub order
        .reset_index()
        .rename(columns={"index": "geo_accession"})
)
expr_sub = expr_sub.set_index(expr_sub.columns[0])
expr_sub.to_csv("/labs/khatrilab/provido/luna_final/LUNA-main/gene_expr_data/final_expr.csv")
expr_sub
```
